```python
import plotly.graph_objects as go
import numpy as np
```

```python
i = complex(0, 1)
s = lambda w: i * w
```

```python
ws = lambda w: ( 10 * (0.05*s(w)+1) * (0.5*s(w)+1) ) / ( s(w) * (0.1*s(w)+1) * (0.01*s(w)+1) * (0.04*s(w)*s(w)+0.03*s(w)+1) )
# ws = lambda w: 1/( s(w) * (0.1*s(w)+1) )
```

```python
W = np.linspace(0, 200, 1001)
WS = 20 * np.log10(np.abs(ws(W)))
# plt.plot(W, WS)
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=np.log10(W), y=WS, name="Lm", line_shape='spline'
        )
    )

fig.update_layout(title='ЛАФЧХ', autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90),
                  template='plotly_dark')

fig.update_yaxes(tick0=0, dtick=20)

fig.show()
```
