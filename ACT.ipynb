```python
def V(x, y):
    R = 1
    r = np.sqrt(x**2 + y**2)
    if r < R:
        return np.array([0, 0])
    else:
        u = np.array([1, 0])
        n = np.array([x, y])/r
        return u - 3*R/4*(u + np.dot(u, n)*n)/r - R**3/4*(u-3*np.dot(u, n)*n)/r**3

xs = np.linspace(-5, 5, 100)
ys = np.linspace(-5, 5, 100)
vx = []
vy = []
for x in xs:
    tx = []
    ty = []
    for y in ys:
        tx.append(V(x, y)[0])
        ty.append(V(x, y)[1])
    vx.append(tx)
    vy.append(ty)

# print(len(xs))
# print(len(ys))
# print(len(np.array(v)[:,0]))
# print(len(np.array(v)[:,1]))

# Create streamline figure
fig = ff.create_streamline(xs, ys, np.array(vx), np.array(vy), arrow_scale=.1)
fig.show()
```
