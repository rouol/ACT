```python
i = complex(0, 1)
s = lambda w: i * w
```

```python
ws = lambda w: 7*(s(w)+1)/(s(w)*(0.01*s(w)+1)*(0.5*s(w)+1) + 7*(s(w)+1))
#ws = lambda w: 10 / s(w) / (0.1*s(w)+1) / (0.08*s(w)+1)
```

```python
W = np.linspace(-1, 3, 1001)
WS = 20 * np.log10(np.abs(ws(np.power(10,W))))
print(20 * np.log10(np.abs(ws(np.power(10, 1.588)))))
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=W, y=WS, name="Lm", line_shape='spline',
        # mode='lines+markers',
        line=dict(color='black')
        )
    )

fig.update_layout(
    # title='ЛАЧХ',  
    autosize=False,
    width=500, height=500,
    margin=dict(l=50, r=50, b=50, t=50),
    template='simple_white',
    xaxis_title=r'lg(ω)',
    # yaxis_title=r'$d, r \text{ (solar radius)}$'
    )

fig.update_yaxes(tick0=0, dtick=20)

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.update_yaxes(range=[-160, 25])
fig.update_xaxes(range=[-1.5, 3.5])

fig.add_annotation(
    x=3.5, y=-160,
    text="",
    ax=-150, ay=0,
    xshift=2,
    yshift=0,
    showarrow=True,
    arrowsize=20,
    arrowwidth=0.1,
    arrowhead=1
    )

fig.add_annotation(
    x=-1.5, y=25,
    text="",
    ax=0, ay=150,
    xshift=0,
    yshift=2,
    showarrow=True,
    arrowsize=20,
    arrowwidth=0.1,
    arrowhead=1
    )

fig.show(), plotly.io.write_image(fig, 'Численная ЛАЧХ для замкнутой системы.pdf', format='pdf')
```
