```python
import plotly
import plotly.graph_objects as go
import numpy as np
```

```python
i = complex(0, 1)
s = lambda w: i * w
```

```python
ws = lambda w: ( 10 ) / ( s(w) * (0.1*s(w)+1) * (0.08*s(w)+1) )
# ws = lambda w: 1/( s(w) * (0.1*s(w)+1) )
```

```python
W = np.linspace(0, 200, 1001)
WS = 20 * np.log10(np.abs(ws(W)))
fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=np.log10(W), y=WS, name="Lm", line_shape='spline',
        # mode='lines+markers',
        line=dict(color='black')
        )
    )

fig.update_layout(
    # title='',  
    autosize=False,
    width=500, height=500,
    margin=dict(l=50, r=50, b=50, t=50),
    template='simple_white',
    xaxis_title=r'lg()',
    # yaxis_title=r'$d, r \text{ (solar radius)}$'
    )

fig.update_yaxes(tick0=0, dtick=20)

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.update_yaxes(range=[-80, 50])
fig.update_xaxes(range=[-1, 2.5])

fig.add_annotation(
    x=2.5, y=-80,
    text="",
    ax=-150, ay=0,
    xshift=2,
    yshift=0,
    showarrow=True,
    arrowsize=20,
    arrowwidth=0.1,
    arrowhead=1
    )

fig.add_annotation(
    x=-1, y=50,
    text="",
    ax=0, ay=150,
    xshift=0,
    yshift=2,
    showarrow=True,
    arrowsize=20,
    arrowwidth=0.1,
    arrowhead=1
    )

fig.show(), plotly.io.write_image(fig, '1.pdf', format='pdf')
```

```python
W = np.linspace(-100, 1000, 1001)
WS = ws(W)
argWS = np.arctan2(WS.imag, WS.real) * 180 / np.pi
argWS[argWS > 0] -= 360

fig = go.Figure()
fig.add_trace(
    go.Scatter(
        x=np.log10(W), y=argWS, name="Lm", line_shape='spline',
        # mode='lines+markers',
        line=dict(color='black')
        )
    )

fig.update_layout(
    # title='',  
    autosize=False,
    width=500, height=500,
    margin=dict(l=50, r=50, b=50, t=50),
    template='simple_white',
    xaxis_title=r'lg()',
    # yaxis_title=r'$d, r \text{ (solar radius)}$'
    )

fig.update_yaxes(tick0=0, dtick=90)

fig.update_xaxes(zeroline=True)
fig.update_yaxes(zeroline=True)

fig.update_xaxes(showgrid=True)
fig.update_yaxes(showgrid=True)

fig.update_yaxes(range=[-270-45, 0+45])
fig.update_xaxes(range=[-1.5, 3.5])

fig.add_annotation(
    x=3.5, y=-270-45,
    text="",
    ax=-150, ay=0,
    xshift=2,
    yshift=0,
    showarrow=True,
    arrowsize=20,
    arrowwidth=0.1,
    arrowhead=1
    )

fig.add_annotation(
    x=-1.5, y=0+45,
    text="",
    ax=0, ay=150,
    xshift=0,
    yshift=2,
    showarrow=True,
    arrowsize=20,
    arrowwidth=0.1,
    arrowhead=1
    )

fig.show()
```
